# What are Infographics? 

An *infographic* is simply a visual used to communicate information and/or data.  An infographic typically goes above and beyond the data visualizations you have learned in R, Python, and Tableau, since they are not tied down to the traditional chart types you have thus far encountered to deal with data. Can you use some of those traditional charts within your infographic? Absolutely! Does the entire visual revolve around a traditional chart? Absolutely not! 

With attention spans getting shorter and shorter, it is critical that you can convey all the important information at a glance.  Perusers of your findings should not have to undergo any intense study to understand what is going on with your data! This is true of all data visualizations, but infographics take it to the next level.  They rely on simple graphics, easy-to-read text, and enough space to make perusing the data a joy, not a chore.

---

## When to Use an Infographic

You can use an infographic to do the following: 

* Illustrate data
* Reduce complexity 
* Compare things
* Raise awareness

---

## Comparing Data Visualization and Infographics

Infographics are a particular type of data visualization, so they share a lot of overlap.  They both seek order, and are both meant to be informative. Further, they both can involve highly skilled people from a large number of disciplines, such as programmers, designers, statisticians, journalists, developers, geneticists, economists, and so forth. Both data visualization and infographics hopefully lead us to the still more refined state: Knowledge.

However, data visualization can be created for the purposes of making sense of the data, or to change a data set into a more accessible form. Often data sets can be so massive and unwieldy that it can be almost impossible or laborious for a human to process. Therefore data visualizations are far more likely to be generated automatically through the use of algorithms or computer programs than infographics, which may use graphic design software. Infographics often contain summaries of the data, or look at things in aggregate (means, frequencies), while data visualizations are more likely to look at all the data in a chart.

Because infographics typically do not include all the data, this means that they're more focused, which can be a boon.  But it also means that they are slightly more likely to inadvertently introduce bias by not presenting the whole picture.  This is not necessarily a bad thing; typically infographics are created for a specific purpose and are used for marketing or public awareness campaigns.  But it is something to be aware of as you peruse a wide variety of infographics.

# Steps for Infographic Creation

Believe it or not, there is a lot of thinking that goes into creating an infographic before you even begin drawing it up. As a general rule of thumb, you want to really have your goals in mind before you set out, or you are doomed to fail.  You need to define the purpose of your infographic, know your audience, establish your key message, and then start translating your data and findings into plain English.  Then, and only then, are you ready to start creating your actual infographic! This wonderful infographic about how to make infographics says it all (whoa, so meta!):

---

## Step 1: Define your Purpose

You probably won't wake up in the middle of the night with a burning desire to create an infographic, and decide to do so in your spare time just for fun.  Instead, infographics are usually designed for a particular purpose. It may be to help with your company's marketing efforts, it may be to communicate findings to upper leadership, it may be to educate a population...whatever the reason, it is important to really hammer out why the infographic is being created.  Without a clear purpose, you are likely to spend a lot of work on something that doesn't get used much.  No one wants that!

---

## Step 2: Know your Audience

Now that you know WHAT you're trying to accomplish, you need to know FOR WHOM you're doing it! Your target audience will determine the language you'll use, the images they'll recognize, and the base knowledge that you'll need to convey. If you don't have a clear picture of who you're trying to reach, you may very well fail to communicate effectively.

---

## Step 3: Establish your Message

What do you absolutely need your audience to know to meet your purpose? That is your message.  You don't have much time or space - most people are just going to glance over this bad boy - so make this message a one point take-away that you are going to hit 'em over the head with! 

Magnolia Consulting provides a **[wonderful worksheet](https://magnoliaconsulting.org/wp-content/uploads/2018/03/Crafting-a-Powerful-Message.pdf)** to walk through on how to create a powerful message.

You are encouraged to use it to walk through the steps as needed.

---

## Step 4: Translate your Data and Findings

Now that you know what you're doing, why you're doing it, and who you're doing it for, it's time to pull in the data that will back it up! An infographic is not the time to vomit data everywhere.  You will be carefully choosing only a couple metrics to highlight that back up your message and resonate with your audience, and you'll want to ensure that you're providing aggregate metrics - raw data has no place in the infographic world.

---

## Step 5: Create your Infographic

The last step is to actually get to work making an infographic! You'll learn more details in the remainder of the lesson on how to actually execute this daring feat, but always remember that if you don't have the planning details down, no matter how beautiful you create your infographic, it may still end up flopping.

# Color 

Before you get into using color, you should know a little bit about it.

---

## Properties of Color

There are two properties of color you should be familiar with - the *hue* and the *saturation* of a color.  Hue is what you think of right away when you think of colors - red, green, blue, etc.  Saturation is how light or dark that color is; it literally is defined as how much white you have mixed in with the paint, if you were an artist. 

Below you have a color wheel.  As you go around the perimeter of the wheel, that is your hue.  As you go from the middle of the wheel out, that is your saturation.  The closer to the edge you get, the more highly saturated your color is.

![A circle is divided into 12 equal parts that are shaded with different colors.](Media/infographic19.png)

---

## How Computers Process Color

There are two main system your computer uses to process colors - the *rgb* system or the *hex code* system. You can think of the rgb system like your print ink.  There are three base colors, a red-ish, a yellow-ish, and a blue-ish, and they can combine to make all the other colors.  The same is true in rgb.  Each color has it's own value from 0-255 that tells you how much of each color you are adding, with 0 being none, and 255 being all. A nice dusky pink, for instance, might be ```255, 153, 204```. These values go in order of the system rgb - so the 255 is for red, the 153 is for green, and the 204 is for blue.

The hex code system uses a six-digit, hexadecimal number to represent color.  Again, there are three blocks of code for red, green, and blue, but instead of a number for each ranging from 0-255, there is now a number ranging from zero to FF. A hex code will always start with a #, and uses the format ```#RRGGBB```. The same dusky pink in hex code is #FF99CC.

You won't need to generate these codes on your own, or even know what they mean just by looking at them, but knowing how color is represented on your computer will allow you greater control of color when you start creating your own infographics and will allow you to match colors if necessary. With over 16 million colors out there, getting back to "that one you used before" can be tricky unless you've saved the code or know how to access it!

# Page 7 - What Colors Should you Use? 

With color, the general rule of thumb is that you don't want any more than three colors or shades of the same color.  Any more than that, and you are distracting the eye and the reader will spend more time taking in your colors than your message! Generally, if you go with three colors, you want a main color, a secondary color, and a third color that is just used a little bit as an accent.  This third small accent color can be something bright and bold - you are using it sparingly, so it can be a little bit further out there.

---

## Basic Color Palettes 

When you start putting together different colors, they form a color palette.  There are several ways that a color palette can be built.  The basic ones are shown below: 

![Four bars shaded in different colors are labeled monochromatic, analogous, complementary, and triad.](Media/infographic18.png)

The first is the *monochromatic* color palette.  This uses all the same hue, but has different levels of saturation. 

The second is an example of an *analogous* color palette. It keeps the saturation constant, but uses different hues.

The third row demonstrates the *complementary* color palette.  If you follow a traditional color wheel, then complementary colors are on the exact opposite side of the color wheel from each other.  This can look a bit jarring at first, especially if you keep the saturation up really high, like the red and green colors all the way to the left.  But if you start to bring down the saturation, like the pair all the way on the right hand side, using a complementary color palette can really look keen.

The last row demonstrates the *triad* color palette.  No, you're not suddenly learning about the Chinese mafia here. The triad is when your three colors are evenly spaced out along your color wheel.

If you're thinking to yourself that this is all a bit complicated, and that as a data scientist you're much more comfortable with numbers than colors, never fear! There are plenty of websites out there that provide you with ready-made color palettes you can utilize. **[Adobe Color](https://color.adobe.com/create)** is one of the better ones, and even provides you with the basic color palette options! From there, it's up to you to move around the color wheel and pick something you find visually appealing. It provides you with the hex codes for each color so you can easily replicate them in your own work.

---

## Accessibility is Key

In all cases, but especially if your infographic is far-reaching or for government and non-profit organizations, you will want it to be accessible! This means that anyone who has a disability can still use and understand your work.  For websites and things, there are lots of laws about accessibility, but for infographics, the number one issue you need to be concerned with is those who are color-blind. There are two types of color blindness - red/green color blindness, in which the person can't differentiate between red and green, and blue/purple color blind, in which the person can't differentiate between blue and purple colors. Here's the traditional color-blind test: 

![A circle filled with shaded small circles represents a pattern of number two.](Media/infographic20.png)

If you were color-blind, you wouldn't be able to see the number two printed within the circles at all, because red and green look exactly the same! And since, in most cases, you want EVERYONE to be able to fully read and enjoy your infographic, you'll want to take this into account.  Luckily, **[Color Brewer](https://colorbrewer2.org/#type=sequential&scheme=PuBu&n=3)** has got you covered - this is a website that shows color palettes with colorblind safe options! 

![A window is displayed in two panels. The left panel has a text field labeled the number of data classes and it is given as three. Next, three radio buttons are placed. The radio buttons are labeled sequential, diverging, and qualitative. Two color palates labeled multi-hue and single hue. Three checkboxes labeled colorblind safe, print friendly, and photocopy safe. Another set of checkboxes is placed. The checkboxes are labeled roads, cities, and borders. Two radio buttons are placed to select the background. The radio buttons are labeled solid color and terrain. The right panel has four tabs on top. The tabs read, how to use, updates, downloads, and credits.](Media/infographic21.png)

Above is what that webpage looks like, and on the left hand side, there are check boxes for not only colorblind safe palettes, but also printer friendly and photocopy safe options! How smart is that? If you know that you are producing an infographic that will be distributed primarily in print, and not on the web, it is strongly encouraged that you choose those options, so that your colors don't end up completely washed out and your message lost with numerous photocopies!

---

## Matching Colors

If you're working for a company, there's a good chance that you will not have the luxury of choosing your own color scheme, but will instead be forced into the company's color scheme or to make your color scheme fit their logo. This can be a bummer if your company has chosen their logo / webpage scheme poorly, but can be a blessing if it was properly designed.  

You can pull either the rgb or the hex code off ANY webpage so that you get an exact match. All you have to do is right click on the image or element, choose the ```Inspect``` option, and then look in the color section for the information. There can be a lot of info, and it can be confusing, though.  So what if there was a better way? What if you could upload an image and instantly end up with the hex codes for it? Lucky for you, **[HTML Color Codes](https://html-color-codes.info/colors-from-image/)** has your back! Simply upload a file, like your company logo, and then you'll be able to click in all the different parts and explore the hex codes!

![Four bars shaded in different colors are labeled monochromatic, analogous, complementary, and triad.](Media/infographic18.png)

The first is the *monochromatic* color palette.  This uses all the same hue, but has different levels of saturation. 

The second is an example of an *analogous* color palette. It keeps the saturation constant, but uses different hues.

The third row demonstrates the *complementary* color palette.  If you follow a traditional color wheel, then complementary colors are on the exact opposite side of the color wheel from each other.  This can look a bit jarring at first, especially if you keep the saturation up really high, like the red and green colors all the way to the left.  But if you start to bring down the saturation, like the pair all the way on the right hand side, using a complementary color palette can really look keen.

The last row demonstrates the *triad* color palette.  No, you're not suddenly learning about the Chinese mafia here. The triad is when your three colors are evenly spaced out along your color wheel.

If you're thinking to yourself that this is all a bit complicated, and that as a data scientist you're much more comfortable with numbers than colors, never fear! There are plenty of websites out there that provide you with ready-made color palettes you can utilize. **[Adobe Color](https://color.adobe.com/create)** is one of the better ones, and even provides you with the basic color palette options! From there, it's up to you to move around the color wheel and pick something you find visually appealing. It provides you with the hex codes for each color so you can easily replicate them in your own work.

---

## Accessibility is Key

In all cases, but especially if your infographic is far-reaching or for government and non-profit organizations, you will want it to be accessible! This means that anyone who has a disability can still use and understand your work.  For websites and things, there are lots of laws about accessibility, but for infographics, the number one issue you need to be concerned with is those who are color-blind. There are two types of color blindness - red/green color blindness, in which the person can't differentiate between red and green, and blue/purple color blind, in which the person can't differentiate between blue and purple colors. Here's the traditional color-blind test: 

![A circle filled with shaded small circles represents a pattern of number two.](Media/infographic20.png)

If you were color-blind, you wouldn't be able to see the number two printed within the circles at all, because red and green look exactly the same! And since, in most cases, you want EVERYONE to be able to fully read and enjoy your infographic, you'll want to take this into account.  Luckily, **[Color Brewer](https://colorbrewer2.org/#type=sequential&scheme=PuBu&n=3)** has got you covered - this is a website that shows color palettes with colorblind safe options! 

![A window is displayed in two panels. The left panel has a text field labeled the number of data classes and it is given as three. Next, three radio buttons are placed. The radio buttons are labeled sequential, diverging, and qualitative. Two color palates labeled multi-hue and single hue. Three checkboxes labeled colorblind safe, print friendly, and photocopy safe. Another set of checkboxes is placed. The checkboxes are labeled roads, cities, and borders. Two radio buttons are placed to select the background. The radio buttons are labeled solid color and terrain. The right panel has four tabs on top. The tabs read, how to use, updates, downloads, and credits.](Media/infographic21.png)

Above is what that webpage looks like, and on the left hand side, there are check boxes for not only colorblind safe palettes, but also printer friendly and photocopy safe options! How smart is that? If you know that you are producing an infographic that will be distributed primarily in print, and not on the web, it is strongly encouraged that you choose those options, so that your colors don't end up completely washed out and your message lost with numerous photocopies!

## Matching Colors

If you're working for a company, there's a good chance that you will not have the luxury of choosing your own color scheme, but will instead be forced into the company's color scheme or to make your color scheme fit their logo. This can be a bummer if your company has chosen their logo / webpage scheme poorly, but can be a blessing if it was properly designed.  

You can pull either the rgb or the hex code off ANY webpage so that you get an exact match. All you have to do is right click on the image or element, choose the ```Inspect``` option, and then look in the color section for the information. There can be a lot of info, and it can be confusing, though.  So what if there was a better way? What if you could upload an image and instantly end up with the hex codes for it? Lucky for you, **[HTML Color Codes](https://html-color-codes.info/colors-from-image/)** has your back! Simply upload a file, like your company logo, and then you'll be able to click in all the different parts and explore the hex codes!

# Choosing a Font

The same general rule that applies for colors applies for fonts - no more than two to three different types on an infographic! Make sure they work well with each other and are easy to read.  Have you ever had to squint to make out a weird, curly font, or something that was just too small? That is NOT what you are going for here. You can have some variety in how you use those 2-3 fonts you've chosen; for instance, you could use italics, bold, or vary the sizing to emphasize certain points, but again, don't go crazy - remember to KISS!

# Flow

The flow of an infographic is very important - you want it to draw the reader in and ensure that they are being navigated through your infographic properly.  You don't want their eyes to randomly flit around; instead you want to ensure that they move smoothly from one section to the other in the order you have decided! How do you make that happen? By controlling the flow! Use visual elements like color, sizing, and font to draw attention and then transition that attention around the page.  

In the below infographic, your eye immediately starts up top, with the header ```Food + Fuel```.  Then you're drawn to the middle, where you see that 80% of soy beans are eaten.  You're then drawn down to the large statistics on the bottom left, since a) they are larger and b) in the English language, you read left to right. All three of these main headers also make use of the same color brown, so you know that they are primary information.  All the secondary information, that is text heavy, is located on the right side of the page, and has a green header.

### Change Usage Settings

First, it is absolutely critical that you have permission to use these images.  The last thing that you want to do is get your company in trouble with copyright law! Google has cleverly developed a button with just this problem in mind.  Click on the ```Tools``` button, then in the middle section where it says ```Usage rights```.  The most stringent, or rigorous, option here is ```Labeled for reuse with modification```.  This means you can reuse the graphic in any way you like! It can be for a business or for your own private use, and you can even change it if you see fit! After that is ```Labeled for reuse```.  This is the filter you most likely want to use if you are working for a company and don't intend to modify the image, since it will cover commercial use.  If you're doing something for personal demonstration, however, your options get a little broader - you can choose ```Labeled for noncommercial reuse with modification`` or ```Labeled for noncommercial reuse```, depending on what your needs are.  Typically, ```Labeled for reuse``` will cover most of your needs. You'll know you've activated a filter when the text for it turns blue and is shown in the toolbar.

## What's Wrong with these Infographics?

This first infographic is very busy! In addition, you have no focal point - the word "libraries" appears three times, so where do you look? Your eye skips right over the blue text at the top, so you land in the "libraries libraries libraries" section next. As you proceed down the page, the numbers are the focal point, not what those numbers represent, so you are left confused and unsatisfied. Lastly, the image in the middle on the very bottom shows a counter-intuitive depiction, with full smileys on the right, and full frownies on the left.  Typically in our culture, you would see this in reverse, so it has the potential to confuse people greatly.
