# Type I and Type II Error Review

You have been working with hypothesis testing for a while, so it's time to review Type I and Type II error rates, and then build upon those concepts. *Type I Error* is when you reject the null hypothesis when you shouldn't have.  This means you reported finding something in your data when there was really nothing.  *Type II Error* is when you accept the null hypothesis when you shouldn't have. This means that you reported finding zilch in your data, but really, there was something interesting of note. 

![A table has two columns and two rows. The column headings are labeled reject null and accept null. The row headings are labeled null true and null false. The row entries are as follows. Row 1, Type I alpha, correct 1 alpha. Row 2, correct, type II.](Media/power1.png)

---

# What is Power? 

Type I error is also known as *alpha*, while Type II error is also known as *beta*. If you've gotten the correct answer by accepting the null correctly, then this is one minus alpha.  But if you've gotten to the correct answer by rejecting the null correctly, then this is conceptually one minus beta, which is also known as *power*.  Power is important, because it determines whether you have enough data to detect an effect.  Just like your phone won't charge without power, a statistical test won't find anything if it doesn't have enough power.  Said another way, power is the probability of rejecting the null hypothesis when the null hypothesis is false (correct). 

Here are some things to remember about power: 

* As the *n*, or your sample size, increases, so does power.
* As the standard deviation of the population increases, the power decreases. 
* Power is greater in a one-tailed test than a two-tailed test.

---

## Types of Power

* **A priori power:** Examine your power ahead of time, so you can gather enough data to get a satisfactory result with your statistical analysis.
* **Post hoc / observed power:** Examine your power after the fact, to see if your conclusions are valid.

---

## Typical Power Value

Just like you can assume that alpha, your significance level benchmark, will always be .05 unless otherwise stated, it is typically accepted that power will be .8 unless otherwise stated.  This means that you have an 80% chance of detecting an effect if it's there.

# Power Analysis for t Tests

There are a wide variety of different tests in the *t*-test family. You will select *t* tests from the drop down of ```Test family``` on the left, and then a list of the different tests will populate in the ```Statistical test``` menu on the right. There are many more tests in the *t*-test family than you might have realized or learned; nor do you need to learn all of them.  

![A window has two panels. The left panel displays two dropdown list boxes labeled test family and type of power analysis. A box labeled determine is placed at the bottom of the left panel. The right panel is labeled statistical test on top of a list from which the sixth option is selected. The sixth option reads, means: difference between two dependent means open bracket matched pairs close bracket.](Media/power10.png)

The ones that align with what you have learned for *t*-tests start with ```Means```:

* **Means: Difference between two dependent means (matched pairs):** Dependent *t*-test
* **Means: Difference between two independent means (two groups):** Independent *t*-test
* **Means: Difference from constant (one sample case):** Single sample *t*-test

You will also find the information for a simple linear regression (only one IV) under the name ```Linear bivariate regression: One group, size of slope```. 

All of the options that have in them ```Wilcoxon``` are *t* tests that can be done when you do not meet the assumption of normality for a regular *t*-test. 

---

## Input Parameters

Once you have chosen the appropriate test, you will then need to define input parameters.  These are the specifications about the analysis you'd like to run. For instance, if you'd like to run an independent *t*-test and have selected the option to do so, you will need to provide the following input parameters, as shown below:

* Tail(s) for your test
* Effect size using Cohen's D
* Alpha level
* Power level
* Ratio of sample size per group

![A window displays three dropdown list boxes labeled test family, statistical test, and type of power analysis. The input parameters field has five fields labeled tail, effect size d, alpha err prob, power of 1, minus beta err prob, and allocation ration N2 by N1. The output parameters field has seven text fields.](Media/power11.png)

---

### Tail(s)

Although using a one-tailed test will provide you with slightly more power, two-tailed tests are used much more frequently.  This is because they require less prior knowledge (you can just guess that your data differs, not in which direction it differs), and they are a little more practical for real-world data science.

---

### Effect Size

You have been briefly introduced to effect size using Cohen's D for *t*-tests.  The important thing to remember about effect size is that they are a better measure of your statistical test than *p* value because they are not based on sample size.  The *p* value is inherently flawed, because it will get smaller the more data you add! By the time you get to a dataset of a thousand or more, you are almost guaranteed to find significance sometimes because of the calculations involved! 

You'll want to plug in a number here for effect size that makes sense for the information you want to analyze.  The standard is typically a medium effect size, but you may want to have enough data to hit someone over the head with your results, which would require a large effect size, or you may be doing something where even a small effect makes a difference - in healthcare, for instance, where even the very small preventive effects of one aspirin a day can help prevent heart attacks! 

If you have forgotten what makes a small, medium, or large effect size for Cohen's D, you're in luck - G*Power has your back! Simply hover over the ```Effect size d``` input box and you will get a reminder: 

![A window displays three dropdown list boxes labeled test family, statistical test, and type of power analysis. A few text fields are placed on the left and right side of the box labeled input parameters and output parameters. The input parameters field has five fields labeled tail, effect size d, alpha err prob, power of 1, minus beta err prob, and allocation ration N2 by N1. 0.5 is entered on the effect size d. A popup message indicating the effect size d box and the message reads, effect size conventions d equals .20 minus small, d equals .50 minus medium, d equals .80 minus large.](Media/power12.png)

---

### Alpha

Just like with anything else, your alpha level will typically be .05 unless otherwise stated or decided. G*Power's default is .05, so you won't have to worry about it.

---

### Power

Generally, you will go with a standard power level of .8.  You will need to change this manually, as G*Power sets the default as .95, which is pretty high. 

---

### Allocation Ratio

The last input parameter is for the ratio of your sample sizes between groups in your independent variable. If they are relatively equal, then you can leave the allocation ratio as the default of 1.  But if you anticipate having very unequal sample sizes, you'll need to change it, so that the ratio mimics the ratio in your data. For instance, if you are comparing differences between men and women in data science, you know that there are far more men than women, so you might want to consider changing this ratio to 2, meaning you expect to collect data from only one women to every two men.

---

The output you receive from your power analysis will show up in two places: the top section, and the right hand side in the bottom under ```Output Parameters```.  When doing an ```A priori``` test, the most important output you can receive is the sample size.  It typically will provide you with both the total sample size needed and the sample size needed for each group, in case you haven't chosen to have them evenly distributed with an allocation ratio of 1.

## Input Parameters for Between-Subjects Designs

Take a look at the ```ANOVA: Fixed effects, special, main effects, and interactions```. There are five input parameters here that you will go over: 

* Effect size f
* alpha err prob
* Power
* Numerator df
* Number of groups 

---

### Effect Size f

Although there are many different effect sizes you can calculate for the ANOVA family, one of the more common is ```f```. G*Power kindly tells you what the small, medium, and large parameters are for f if you hover over the cell: 

![A window displays three dropdown list boxes labeled test family, statistical test, and type of power analysis. A few text fields are placed on the left and right side of the box labeled input parameters and output parameters. The input parameters field has five fields labeled tail, effect size d, alpha err prob, power of 1, minus beta err prob, and allocation ration N2 by N1. 0.5 is entered on the effect size d. A popup message indicating the effect size d box and the message reads, effect size conventions f equals .10 minus small, f equals .25 minus medium, f equals .40 minus large.](Media/power15.png)

As with *t*-tests, you will want to pick an effect size range that makes sense for the task at hand. The default medium effect size of .25 should do just fine for most things, but if it's really important that you detect a large effect, you'll probably want to go with something a little larger, and if even a small effect can have a large difference, like a reduction in violence or in the prevention of disease, then you'll want to make your f value smaller.

---

### Alpha Error Probability

Like with all tests, you can keep this default at alpha = .05 unless otherwise specified.

---

### Power

Just like with the *t*-tests, don't forget to change your power from G*Power's default of .95 to a more reasonable .8, which is typically the standard for power.

---

### Numerator Degrees of Freedom

The numerator degrees of freedom is just the number of variables you plan to have in your model minus one.  So, if you were going to have a two-way ANOVA, this would be 1.  If you were going to have a 4 way ANOVA, the value you would place in ```Numerator df``` would be 3.  You'll note that you won't see this box pop up if you select the ```ANOVA: Fixed effects, omnibus, one-way``` because by default, you are only including one variable.

---

### Number of Groups

This is simply the number of levels of the main independent variable you intend to test.  Trying to determine how five age groups differ? Then you'll place 5 here.  Examining the difference between three dog breeds? Then go ahead and enter a 3 instead.

---

# Input Parameters for Repeated Measures Designs

With a more complex design comes even more input parameters! In addition to those you just learned about, for ```ANOVA: Repeated measures``` you will also need to include information about: 

* Number of measurements
* Corr among rep measures
* Nonsphericity correction

These parameter guidelines apply to anything with ```Repeated measures``` in its name, including ANOVAs and MANOVAs.

---

## Number of Measurements

The number of measurements is simply the number of timepoints (or other repeated measurement) you'll be examining.  If you have a pre and a post test, then this value will be 2.  If you are looking at sales data every quarter for a year, this number would be 4.

---

## Correlation Among Repeated Measures

The correlation among repeated measures is the average correlation value, *r* of how each repeated measure relates to each other.  If you remember back from your correlation interpretation values, .5 is a moderate correlation, exactly in the middle.  You typically will need to use your best guess here to determine what an appropriate *r* value might be, since most likely you don't have access to your data yet. If thinking through the issue says that your measures will be really closely related, then you'll probably want to bump these values up, but if you're thinking they might not relate much, then you'll probably want to bump that default .5 value downwards.

---

## Nonsphericity Correction

You can leave this default as 1.

## Summary

Statistical power is important, because it provides you the necessary oomph to ensure you find the correct results when doing analyses. You have now learned how to calculate power for most tests you will run "in the wild." Typically, you will most often calculate power before you begin analyzing data, rather than after you have already analyzed the data.
