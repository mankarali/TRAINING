# What is a Survey?

A survey is a way to gather information from others in which they are asked to provide answers to written questions.  The overarching goal of any survey is to translate a concept into some measurable data that can be studied and analyzed.  This presents many problems and can introduce a lot of biases, but with a solid foundation in survey design, you will hopefully be able to evade a lot of the common traps. 

---

## Survey Design Steps

Like anything else even somewhat creative, it pays to not rush in blindly, but rather to spend some time thinking through requirements and figuring about your overall goals.  Here are the general steps in survey development: 

* Determine the overall purpose of the survey
* Brainstorm the questions you are trying to answer with the survey
* Create survey items that align with your questions
* Format the survey as appropriate for your questions and your audience
* Write scoring and administration rules if needed
* Test out your survey
* Administer your survey

---

## Determine the Overall Purpose

The overall purpose will typically be set by your supervisor or client.  It could be something broad, like "See if employees are happy" or it could be something a little more specific, such as "See if there are any differences in employee satisfaction by department."  Either way, do your best to find out exactly what they're looking for, by asking as many questions as possible.  Especially if you have to start with something broad, you'll want to drill down to exactly what your supervisor or client wants.  In the case of "see if employees are happy," you might start by asking how they would define happiness and if they want to look at any sub-categories of employees.  The answer to these questions will start driving you in the right direction.

---

## Brainstorm Questions to Answer

Once you have a very broad, overall purpose, you'll want to get into even more specifics. What are the questions you absolutely MUST be able to answer with this survey when you're done? Typically these are called *evaluation questions* or *research questions*, and it's a good idea to have at least three, though they may have subparts to them.  For instance, if you are surveying caregivers of the elderly, one of your overarching evaluation questions might be "What resources do caregivers need to best care for their care recipient?" and you can answer that question by looking at physical, mental, and social needs. 

Take your time brainstorming these questions, and if possible, work with the supervisor or client who set your overarching goal to make sure you're getting at the things he or she cares about most.  These evaluation questions will guide all of the survey items you create, as well as your survey audience.

In the next few pages, you will learn about the next step in the process, which is to create survey items that align with your questions.

# Question Types

There are many different ways you can ask questions on a survey. Some will work better than others for your intended purpose, so it's a good idea to understand the different types so you can make the best use of your survey!

---

## Multiple Choice

You are probably pretty familiar with the multiple choice question, since it is the item of choice on many a school exam.  Here is an example of how a multiple choice question could be used on a survey: 

```text
What is your approximate yearly income? 
-Less than $15,000
-$15,00-$29,999
-$30,000-$59,999
-$60,000-$89,999
-Above $90,000
```

With a multiple choice question, there is only one correct answer, which means that all the options must be exclusive - you shouldn't have any survey respondent who fits in more than one income category, for example. Multiple choice questions provide you with categorical data.

---

## Select All that Apply 

A *select all* question is very similar to a multiple choice question, but instead of having only one correct answer, there can be many. These questions typically start or end with the phrase "Please select all that apply," which is how they have earned their name. Here is an example of a select all question: 

```text
Which of the following items did you spend money on in the last month? Please select all that apply.
[] Gas
[] Groceries
[] Rent / Mortgage
[] Eating out
[] Hobbies
[] Clothes
[] Travel
[] Utilities
[] Household goods
```

Just because someone has spent money on gas in the last month does not mean that he or she could not have spent money on any of the other categories, such as rent or groceries.  Therefore a select all question is much more appropriate than a multiple choice question in this scenario, as select all questions are not exclusive, like multiple choice questions are. The data you receive from a select all question is also categorical, but beware! You will not be able to calculate frequencies for the question as a whole, like you might for a multiple choice question, but instead, will have frequencies for each answer choice, since otherwise you would end up with a number of responses that is much greater than the number of people who actually answered the survey!

---

## Scale

Often, surveys are used to measure subjective things such as attitudes. Sometimes, attitudes are measured with multiple choice questions that have responses like "agree" and "disagree," but they can often be done as a scale question.  Scale questions are laid out horizontally and typically have some sort of scale or ruler system to mark out the different response options, like this: 

![A statement that reads, my manager considers my feelings. To the right of the statement is a five point rating scale. The point on the left is labeled strongly agree. The point in the middle is labeled neutral. The point on the right is labeled strongly agree.](Media/survey2.png)

The benefit of a scale question over a multiple choice question when asking about subjective information is that it is more visual, and it takes up less vertical space.  However, scale questions are not as good for those taking surveys on mobile devices such as a phone or tablet, since it can be hard to see all the options at once.

Scale questions also provide you with categorical data.

---

## Matrix

A matrix question is very similar to a scale question, but now, there are more than one! Matrix questions can look like the one below, or they may start with a common question stem, such as: "In the workplace, I feel that..."

![Statement one, my manager considers my feelings. Statement two, my wellbeing is a priority for my manager. Statement three, my manager defends me from unfair criticism. To the right of each statement is a five point rating system. From left to right, the points are labeled strongly disagree, disagree, neutral, agree, and strongly agree.](Media/survey3.png)

Like scale questions, the major benefit of matrix questions is that they save a lot of vertical space.  Another advantage of matrix questions is that they increase organization on your survey, reduce the amount of repetition, and generally help reduce confusion.  However, as with scale questions, matrix questions may not read well on mobile devices.

Scale questions provide you with categorical data only.

---

## Write-in Response

Write-in responses are not used as often as some of the other survey questions above you have learned about; however, they are one of the only ways you can be provided with continuous data, which adds value! For instance, you could have a write in question about age, instead of using a multiple choice question about age categories, like this: 

```text
What is your age? [             ]
```

Write in questions do not always yield continuous data, however; they often will yield qualitative, unorganized string responses instead.  For example, you could ask this: 

```text
What is your favorite food? [          ]
```

And you should be prepared for all chaos to break loose! Write-in responses are a good way to gather continuous data when you need it, and they are also good to employ if you don't know a lot about your target subject or audience and are worried that there are things "you don't know you don't know." It allows your survey respondents to have a little more control over their answers, and you may glean some valuable information you would have missed otherwise if you had restricted their responses in a structured question.  However, the data wrangling that will come with a write-in response can be nightmarish! Even for a simple question about age, unless you put character limits and data type limits on it, you may end up with people typing out numbers, writing silly things like "60 years young," or making typos like "1111 years old," which can't possibly be correct.  So you'll want to weigh the pros and cons of write-in responses very carefully!

For this reason, many data scientists will include write-in response boxes as part of select all that apply questions that say "Other," and allow folks the ability to write in a response only if they don't think the others are adequate and inclusive of their situation.  This limits the amount of data you'll need to sift through, but still allow some question structure.

Write-in responses can provide continuous data or qualitative data, that can be coded into categorical data if needed and possible.

---

## Open-Ended Essay

Open-ended questions are very similar to write-ins, but they allow for much longer responses.  Typically, they will ask deeper questions that require an expansive response.  An example of an open-ended essay question might be: 

```text
Is there anything else you would like us to know about your experience with Data Tech?
```

As might be expected, analyzing these kinds of questions can be a nightmare, and really requires in-depth natural language processing or qualitative analysis to get meaningful data from these questions.  Therefore, they are typically used sparingly.  However, even if you know you won't have the manpower to analyze absolutely everything, including one open-ended question at the end can make the respondent feel included and yield some surprising information you might not otherwise have gained.

Open-ended essay questions only provide you with qualitative data, and it is very difficult to code it into categorical data.

# Sources of Survey Questions

Now that you've learned about the different types of survey questions, you may be overwhelmed at the prospect at writing all those questions up yourself.  But never fear! You can often borrow survey questions from other places legally, and not only is this not considered cheating, but it is encouraged! Don't re-invent the wheel - there's a good chance that someone else has not only created the survey question you are looking for, but has done it far better than you ever could have.  There is a process called *survey validation* that ensures that your survey will be as unbiased as possible, and whenever possible, you want to look for validated survey items so that you don't have to go through this process yourself, as it is quite time consuming and mathematically advanced.  

Where do you find these miraculous validated survey questions? By and large, it will take a little bit of work and some digging, but it is worth it, because you'll be learning a lot about your topic along the way.  For a few quick wins, check out these resources: 

* **[UK Data Service Variable and Question Bank](https://discover.ukdataservice.ac.uk/variables)** : This website allows you to search for survey questions by keyword. 
* **[Constant Contact Sample Survey Questions, Answers and Tips](https://www.constantcontact.com/aka/docs/pdf/survey_sample_qa_tips.pdf)** : This PDF provides survey question suggestions about basic business topics, such as satisfaction with customer service, events, membership, and websites. 
* **[Wikimedia Learning and Evaluation Question Bank](https://meta.wikimedia.org/wiki/Learning_and_Evaluation/Question_Bank)**: This question bank is excellent for demographics and program participation questions.  If you're doing data science for education, you'll be able to use other sections from it as well.
* **[SurveyLegend](https://www.surveylegend.com/likert-type-scale-responses-examples-with-examples/)**: This website provides you with a wide variety of scaled response options for questions, commonly called *Likert Scales*. 

---

## Peer-Reviewed Articles

You can also search for peer-reviewed articles that have included survey questions, and these are often the best source of survey questions, because you can guarantee that they have been thoroughly reviewed, not just by the authors of the article, but by the scientific community at large.  Often, academic researchers will take years to develop their surveys and ensure they are properly validated, so if you have the time to devote to finding them, and your company values rigor, then searching for research articles with the right surveys is absolutely the way to go.

---

### Where Do you Find Peer-Reviewed Articles? 

If you can link into any large online library system, that is the best way to find peer-reviewed articles.  You may be able to get access through your public library, your new workplace, through a previous education experience, or maybe even through a family member going to school that is willing to share their library time.  If you do have access to an online library system, then some of the bigger databases you'll want to search in for validated surveys are as follows: 

* **Academic Search Complete:** Good general topic search database
* **Business Source Complete:** For business topics
* **CINAHL:** For healthcare
* **Cochrane Library:** Healthcare
* **Education Database:** For education
* **Health and Psychosocial Instruments (HaPI):** Provides surveys for health, business, and psychology
* **JSTOR: The Scholarly Journal Archive:** General topic search database
* **MEDLINE:** Healthcare
* **Mental Measurements Yearbook:** Provides validated surveys for psychology and human interest topics
* **PsycINFO:** Articles on behavioral sciences and mental health
* **ScienceDirect:** General purpose science database

In your particular area, there will definitely be additional databases that are helpful and interesting.  Use your own subject matter expertise and good judgement to navigate and find those!

You may be wondering why so many of the databases above are psychological in nature.  This is because a lot of the topics covered in these psychology databases are broad and apply to many other topics as well, such as business and manufacturing (industrial/organizational psychology), human nature (personality, behavior, motivation, etc.), health (health and medical psychology), education (educational and cognitive psychology), and user experiences (neuroscience), to name just a few, all have their backbone in psychology. So don't discount the resource immediately just because it says "psychology!"

# Designing Survey Questions

If after careful searching, you can't find what you want or need, you will need to create your own survey questions. There are some definite rules and best practices when it comes to creating your own survey questions, so focus up! You'll learn best practices for: 

* Questions
* Answers
* Things to avoid

---

## Phrase All Items as Questions

All items on your survey should be phrased as a question, so that there is no confusion! For instance, if you want to know someone's weight, ask "What is your current weight in pounds?" and leave a write-in response blank, rather than just putting "Weight:" with a write-in response blank.  This helps alleviate confusion and makes things much more specific.  you don't want to leave it up to their interpretation, or you might get some odd data.  Maybe they include the weight they'd like to be.  Maybe they include the weight of their pet or child.  You never know what is going on inside a survey participant's head!

---

## Don't Give the Answer You're Looking For in the Question

You'd be surprised how easy it is to accidentally give away the answer you're looking for in the question text. For instance, what if you asked this question: 

```text
Are you aware that data science is a new field?
-Yes
-No
```

By asking this question, you are telling your participants that data science is a new field! So a much better question would be something like this one: 

```text
How did you first learn that data science is a new field? 
-From a friend
-From a coworker
-Social media
-I had not heard of data science before taking this survey
```

Now you have your bases covered, and you are learning much more from the participant, still with a single question.

---

## Don't Make Them Do Math

Anytime you're asking a survey respondent to calculate something is usually a no-go! At best, it is time-consuming and complicated.  They may or may not actually do what you asked.  At worst, they will make a mistake in their calculations and you will end up with inaccurate data.  For instance, don't ask someone what their body mass index is.  That's not a figure anyone is likely to have memorized, and they probably won't know the formula for calculating it.  It's much better to ask for their height and weight, which is something that most do know, and then calculate BMI later.  You'll be able to calculate it easily all down the column using Python and R, and will save them much confusion, muddling, and wrong-answers.

Here's another example: instead of asking for their distance away from your particular company, you could just ask for their address instead and put it into Google Maps later to find out the distance in miles.  Although time consuming, it is bound to be much accurate!

---

## Be Specific

Specificity is key! You've already hit on that a little with defining units, but it also works for defining terms and being clear in your question wording. 

What if you asked someone not very familiar with data scientist how often they did data wrangling? They might be scratching their heads forever, and you probably wouldn't get great data out of it.  This would be a much better option: 

```text
How often do you perform data wrangling skills such as adding columns, flipping data, subsetting data, splitting data, or recoding data? 
```

Now participants will have a much easier time understanding what you are asking and you have defined it nicely.  You could even ask about each example individually if it added value towards your overall survey goals.

## Provide Time Frames

Your metrics need to be time-bound, and your survey questions should be too.  Research has shown that memory recall is really only accurate going back two weeks in history, so try to keep a reporting window short.  Asking someone how many times they downloaded data in the last year just won't work! As a data scientist, you probably have downloaded a lot of data, and it's such a basic thing that you don't keep track of it very well, let alone over the course of a year! But if you can ask how many times they downloaded data that day or that week, it will be much easier to remember and thus you will increase the accuracy of your responses.

# Designing Survey Answers

There are just as many things to take into consideration when designing survey answers, also called *responses*, as there are when writing questions.  

---

## Provide An Exhaustive List of Answers

Ever take a survey and end up frustrated, because there was no way you could appropriately express yourself? Maybe the response options didn't accurately reflect your experiences, or there was no place to write in an "other" experience? That scenario is exactly what you want to avoid in survey creation! So it is important to include an exhaustive list of options.  Here's an example of what NOT to do:

```text
What do you want to do with your hair?
-Make it longer
-Make it shorter
```

In the example above, there are only two options listed for what someone could do with their hair: they could make it longer or shorter. But what if they liked the length of their hair, and wanted to keep it the same? What if they don't pay any attention to their hair and don't have any particular hair-related goals? Then the respondent would be left feeling very frustrated! A better question, with more response options, might be something like this: 

```text
What do you want to do with your hair?
-Make it longer
-Make it shorter
-Maintain my hair length
-I am not trying to do anything with my hair
```

---

## Ensure Answers are Exclusive

Especially when you ask about categories of something, it becomes easy to include answers that overlap accidentally.  Think about this example: 

```text
How old are you? 
-0-25 years old
-25-40 years old
-40-55 years old
-55-70 years old
-Older than 70
```

See anything wrong with it? How about if you're aged 25, 40, or 55? Which option should you choose? You would fall into two categories, and thus when it came time for analysis, you wouldn't be able to confidently say how many people were in each age group.  A much better option would be something like this:

```text
How old are you? 
-0-25 years old
-26-40 years old
-41-55 years old
-56-70 years old
-Older than 71
```

Now, there is only one option for each person, and all the answers are exclusive.

---

## Make Units Clear

It's also incredibly important to make the units you are asking about clear! What if you were asked how long you'd be in a particular job, like this: 

```text
How long have you been with DataLove Corporation? 
-0-6
-7-12
-13-18
-19-24
```

Without any specification on there, that question could easily be interpreted as either months or years.  To get the answers you're looking for, and not bias your data, it's important to specify the unit here.  Correcting that question would look like this: 

```text
How long have you been with DataLove Corporation? 
-0-6 months
-7-12 months
-13-18 months
-19-24 months
```

Since you're building skills on top of each other, you'll notice that none of these categories overlap, either, so the answer choices remain exclusive.

---

## Scale Responses Should be Balanced

When using scales, it's important that the response options are balanced and have equal intervals between the choices.  For instance, something like like is a good choice: 

```text
Strongly agree
Agree
Disagree
Strongly disagree
```

The above works nicely because you have balance - there is a strongly agree at the top, and a strongly agree at the bottom, and agree to counteract disagree in the middle.  This means that a respondent isn't influenced by verbal clues to lean their response one way or the other.  But consider something like this: 

```text
Strongly agree
Agree
Somewhat agree
Disagree
```

Notice how three of the four options are now positive and in some form of agreement? This unconsciously leads the participant to pick one of the top three options, which of course can bias your results.

---

## Use Ranges When Asking About Sensitive Things

Sometimes folks can be put off by having to write out specific information that is sensitive in nature.  For instance, salary, age, or household income may all be something you squirm about when discussing. It becomes a more manageable and approachable topic when it's in ranges.  You don't feel quite so singled out.  However, you might not be able to get that continuous response you were hoping for; so it's a balancing act. Instead of just asking outright something like "What is your household income?" you may instead want to ask the following: 

```text
Into what range does your current household income fall?
-Less than $25,000
-$25,000-$49,999
-$50,000-$74,999
-$75,000 or more
```

---

## Think Carefully About "I Don't Know" and "Neutral" Response Options

If you allow humans to waffle, they will - it's in their nature! So typically, you won't want to use "I don't know" as an answer choice, or use a neutral option in your response scale.  It's better to have them pick a side, even if that side is only "slightly agree" or "slightly disagree."  If you do include a middle choice - be aware of the ramifications.  How impressed is your boss going to be to hear that over half of participants felt neutrally about a particular topic? Have you learned anything useful that can guide actions? Probably not.

The only time you may want to include a "get out of jail free card" is when a survey is mandatory.  As a general rule, a survey SHOULD NOT BE MANDATORY! It goes against ethics and human rights to require that someone participate in a survey, but in a few situations, like when working with Army commanders, there may be extenuating circumstances that require everyone in a unit to participate. The Army calls this "voluntold." If that's the case, you want to ensure that answering all the questions on the survey is not mandatory, and if participants don't feel comfortable answering, they don't have to; or if they do have to answer, they can choose a neutral option or a "I don't know" option so they are not forced to share personal information if they don't want to.

## Avoid Leading Questions

A leading question is a question in which you have accidentally biased the respondent towards answering in a certain way.  For instance, if you asked, "To what extent do you agree with the following statement," you've only said the word agreed, not disagreed, so respondents are going to be slightly biased towards agreeing.  A better way to phrase something like that would be: "To what extent do you agree or disagree with the following statement." 

You often see leading questions in political debates and the like.  You'll hear questions that start with "Wouldn't you agree that...?" or "Don't you dislike...?" Unless you're really thinking on your feet and ready to be assertive, sometimes it's easy to just say "Yes" to whatever the question might be. 

---

## Avoid Double Barreled Questions

A double-barreled question is one in which you ask about two different things in the same question. For instance, the question below asks: 

```text
Do you believe data scientists should be paid more for working remotely and for working with big data?
```

That covers two items: working remotely, and working with big data.  A participant could feel strongly about one part of that, but not the other, and then they would be unsure of how to answer this question.  Chances are, the participant would make some sort of compromise, and you wouldn't get a true answer about either.  This would be much better as two separate questions, as follows: 

```text
Do you believe data scientists should be paid more for working remotely?
-Yes
-No

Do you believe data scientists should be paid more for working with big data?
-Yes
-No
```

--- 

## Avoid Value Laden Terms

A value-laden term is anything that casts subtly casts judgement.  Words like "good", "bad," or "alright" are all value-laden terms.  They aren't specific enough to do any good in a survey question, but they can isolate people.  What if you eat at your desk regularly because you get a short lunch break, but encounter the survey item down below? 

```text
Please rate your agreement or disagreement with the following statement: 
"People who eat at their desk are disgusting crumb monsters."
```

How are you going to feel about that survey item? You'll probably react pretty negatively, especially if you are a neat eater and tidy up after yourself every day.  You may be so angry you stop taking the survey altogether.  You may continue with the survey, but see so much red that you answer all the questions differently than you might normally.  All of that emotion is for nothing, too, because using the term "disgusting crumb monster" doesn't actually provide a lot of specific information. You can't probably do a lot with the responses from that question.  But changing it to something like the below would provide useful information and hopefully not offend any participants: 

```text
Please rate your agreement or disagreement with the following statement:
"It is easy to accidentally get food stuck in my keyboard when I eat at my desk."
```

Now you may actually be able to collect useful information about a topic, and there's no swirl of emotion around the question.

---

## Avoid Double Negatives

Don't you think you should not do this? 

Joking aside, a double negative means that you have two negative statements or "not" statements in a sentence, and they cancel each other out to mean the opposite.  They can be hard to think through and answers can easily become incorrect.  Double negatives located in everyday life are tricky; double negatives in a survey with agree/disagree statements can be disastrous! 

Try this bad boy out: 

```text
To what extent do you agree or disagree with the following statement?
"I do not think that data privacy is unimportant." 

-Strongly agree
-Agree
-Disagree
-Strongly disagree
```

Wahoo! You have to figure out what the statement means before you can figure out if you agree with it or not. This statement means that data privacy is important, because the double negatives cancel each other out.  So, hopefully y'all would chose strongly agree.  But think about how much head-scratching that took! It would be much easier to do something like this: 

```text
To what extent do you agree or disagree with the following statement?
"Data privacy is important." 

-Strongly agree
-Agree
-Disagree
-Strongly disagree
```

Nice and clean and simple now, yes?

Since you're reinforcing your previous learning, note that the response options for this question are once again nice and balanced!

# Survey Formatting Best Practices

Along with writing your questions properly, there are also a set of best practices for how the survey as a whole should be formatted. 

---

## Survey Length

In an ideal world, your survey should take about five minutes for someone to complete.  However, that may be impossible to manage, depending on your goals and upon the wishes of meddling bosses and clients.  Do your best to keep the survey length under 15 minutes, or you'll have a lot of people stop halfway through or plain old not start your survey in the first place.

---

## General Design Principles

The same basic design principles that you've learned about data visualizations and infographics also apply here.  You'll want to keep to no more than 2 or 3 colors and fonts, and avoid neon colors at all costs! The font size should be readable, with it neither being too small (hard to read on a large desktop monitor) nor too large (hard to scroll through on a mobile phone). It should be easy to differentiate the sections of the survey with headers at the top of the page, and you should be able to tell the questions from their answers.

---

## Reading Level

A general rule of thumb is that your survey should be written at a 6th grade reading level.  This means no large, obnoxious long words, and include definitions for things even remotely complex.  You can check the reading level in MS Word by going to ```File -> Options -> Proofing -> Show readability statistics```, and then run a spelling and grammar check.  But there seems to be a few bugs in the current version, so if that's not working for you, or if you don't have access to Microsoft products, there are also multiple websites dedicated to checking reading level online.  Here's one: **[the Readability Test Tool](https://www.webfx.com/tools/read-able/check.php)**.  Simply choose the option to ```TEST BY DIRECT INPUT``` and then copy and paste your text into the box above and press the ```CALCULATE READABILITY``` button. 

![The readability test tool, a website that checks a texts reading level. The website allows you to enter or paste text by direct input. You can also test text by U R L or by referer.](Media/survey9.png)

You get access to many different statistics as a result, but the most important thing to note is the average grade level shown in the white ```TEST RESULTS``` box, as well as the ```Flesch Kincaid Reading Ease``` score.  This score ranges from 0-100, with 0 being the most difficult to read, at a college graudate level, and 100 being the easiest to read, at a fifth grade reading level.  Since you are aiming for a 6th grade reading level or so for surveys issued to the general public, you are looking for a score of above 80. Clearly the example text run through here would not be appropriate for surveys!

![Test results after text is analyzed by the readability test tool, a website that checks a texts reading level. The test results summar reads, your text has an average grade level of about eleven. It should be easily understood by sixteen to seventeen year olds. Readability indices include flesch kincaid reading ease, fifty six point seven, flesch kincaid grade level, nine point nine, gunning fog score, twelve point eight, smog index, nine point four, and coleman liau index, eleven point five.](Media/survey10.png)

---

## Informed Consent Page

Your company may require an ethical review to ensure that your survey will do no harm to participants.  That can be something formal, called the *Institutional Review Board*, or IRB, or it could be something a little less formal.  Whether an ethical review occurs or not, it is important to ensure that you get consent from all participants who are going to take your survey.  You want to make sure that everyone is a willing participant and knows that they don't have to take your survey and that there will be no negative repercussions if they choose not to take your survey.  In order to make sure they can consent properly, participants need information about what you're doing and why.  So it's important to provide all the information possible about your survey is as concise a manner as possible.  This is usually done with a short descriptive paragraph and then some frequently asked questions.  Here are some things to include on your informed consent page: 

* Basic information about your survey and its goals
* Who to contact if participants have questions
* What are the risks/harms that may come from taking the survey?
* What are the benefits of taking the survey?
* Will the participants' information be shared?
* Can the participant stop taking the survey at any time?
* Is participating in the survey voluntary?

Once the participants have all the information they need to make a decision, then there needs to be some action that takes place that serves as their consent.  If you are doing a paper-and-pencil survey, then this action is a signature.  But in the case of an online survey, language like this is typical: 

```text
I understand that my participation in this research study is entirely voluntary.  I may refuse to participate or stop at any time.  

By clicking NEXT, you consent to take part in this study.
```

The participant may choose to stop there, but you know that if they move onto the next page with the survey, then they are willing and understand everything that taking the survey entails.

---

## Demographics

Next, you'll most likely want to collect demographics from your participants - things like age, gender, race, income, location - whatever you think may play a role in the information you're gathering that you'd like to control for later in analyses. Remember that demographics can be sensitive to some people, so it is a good idea to ask them as ranges rather than a write-in response, with the possible exception of age, just because it is so valuable to have it as a continuous number rather than as a category.

---

## Concepts Should Flow

Next, you'll want to include your regular survey questions, and it is important that they are nicely introduced and have a flow to them.  This not only makes it easier for the participants to understand what is being asked of them, but also increases their recall.  When things flow nicely and make logical sense, your brain spends less time processing those details and can spend more time retrieving the information you asked for accurately. 

---

## Survey Logic

There may be survey questions in which if respondents answer one way, there are certain questions that need to be asked next, but if they answer another way, you'd ideally like to ask them different questions.  This is called *survey logic*, or *skip logic*.  Survey logic is incredibly useful for personalizing the survey experience and ensuring that no one has a longer survey than they need. Here is an example: 

```text
1) Would you be interested in a data scientist training program? 
-Yes
    IF YES: Go to question 2
-No
    IF NO: Go to question 3

2) What would be the ideal length of a data scientist training program? 
-1 hour
-2 hours
-3 hours

3) Would you be interested in a software engineer training program?
-Yes
-No
```

It would be silly to have everyone answer all three questions, because if someone is not interested in a data scientist training program, they probably have no input on how long it should be, and even if they do have input, you probably don't want it, because they are not the target audience.  In the example above, the logic is just roughly blocked out on paper, but if you are doing an online survey, it is easy to convert this into a system in which the participant does not even know they are missing questions, and doesn't have to worry about manually skipping questions.

Blocking your survey logic out ahead of time, before going into any software design program, is helpful as a double-check.  You'll spend a lot of time programming survey logic in, and you don't want to make it up on the fly.  

---

## Thank Them!

Once you have finished the survey, it is important to thank your participants! They took time out of their busy day to help you and your company out! Then on the next line, it's always a good idea to reiterate who to contact if they have any questions, now that they have completed the survey.

---

# Survey Review and Testing

Once you think you have a good survey, you'll want to review it extensively to ensure it meets your needs and that it will not harm anyone.  Especially when you are trying to survey a large chunk of the population, there will be no putting the cat back in the bag once it's out! Any mistakes you make will remain. So test, test, test ahead of time!

---

## Review in Detail

The first "test" you will want to do is just to review it yourself in excruciating detail.  Next to things like resumes or article publications, a survey is most likely one of the most public things you will undertake, and you want to put your best foot forward. So check, spelling, grammar, design, flow, and any skip logic very carefully yourself. Then hand it to a coworker, friend, or family member for another look.  If that's not an option, read it backwards or out loud, both of which make your brain think about writing in a different way so you can catch typos and other mistakes.

---

## Have Experts Review

Once you've thoroughly reviewed the basics, now it's time to get a subject matter expert (SME)'s opinion on your survey.  They'll be able to review it at a high level to make sure that you are covering all the concepts and that your response options are thorough.  An SME's opinion is invaluable, so do not skip this step!

---

## Cognitive Interview

The remaining testing items on this page go over types of testing you can do with a small number of actual participants that represent the population you are interested in.  The first is *cognitive interviewing*, which is when you give a respondent the survey and ask him or her to think aloud as they answer the survey questions.  The purpose of cognitive interviewing is to understand what the participant is thinking and feeling as he or she goes through the survey.  This will help you identify any issues with inappropriate questions and responses.

---

## Behavior Coding

The next type of test that can be done is *behavior coding*.  When utilizing behavior coding, you will observe a participant as he or she takes your survey to see if there are any behaviors that might indicate trouble.  Does the participant put his or her head down on the desk in despair? Is there lots of frowning and grumbling? What about biting the nails and other nervous tics? If you can record what this behavior is and on which sections of the survey it is occurring, you can identify any problematic areas that might need to be revised. Behavior coding is particularly useful in checking to make sure that your survey does not cause undue distress.

---

## Response Latency

*Response latency* is the process of timing how long a respondent takes to answer each survey question.  The longer the amount of time it takes a participant to answer, the more likely that the question is trouble.  It may indicate that the question is confusing, does not have the right response options, or is emotionally troubling in some way.

---

## Respondent Debriefing

*Respondent debriefing* is probably the most commonly completed survey testing.  It's simple to ask participants to take the survey, and then give feedback on the experience and the questions. In many ways, the testing respondents can serve as an SME as well to review that you have the right questions and response options. You get a lot of valuable information with very little fuss.

---

## Vignette Analysis

All of the above testing requires the participant to take the survey exactly how they are.  However, *vignette analysis* asks survey respondents to answer the survey "as if" they were in a hypothetical scenario. So things like, "respond as if you had a disability," or "respond as if you were pro data integrity," etc.  would be asked, so that you could get a feel for how a wide range of participants would respond.  This requires a lot of thought on the part of the participant, however, and usually is only done when you can't get other representatives of the population to test.

# Distributing Surveys

Now that you have drafted and tested your survey, it's time to distribute it and gather the participants you need to take it! You'll want to think through the best ways to reach your participants with your supervisors and/or clients.  Possible ways to distribute a survey include: 

* Email
* Slack
* Host a link on an internal website
* Social media sites such as LinkedIn, Facebook, and Twitter
* Provide links via QR codes on posters and pamphlets in locations your participants commonly visit
* Provide paper copies in locations your participants commonly visit

---

## Response Rates

*Response rate* is the number of people who will click on your survey link.  *Completion rate* is the number of people who will actually finish taking your survey.  Unfortunately, these numbers are not often the same, and you will lose some people taking your survey at the consent page, as they learn more about what will be asked of them, as well as losing some people in the middle of the survey, when it gets too long, too boring, or they otherwise get distracted.  A good survey response rate is about 15%, though of course more is always desired! This means that as you determine who to reach out to, you will need to reach out to many more people than you actually need in order to collect an adequate sample.  For example, if you sent a survey to 100 people, with a survey response rate of 15%, you would only get 15 people.  So, if you did in fact need 100 people, you would need to reach out to 667 people. 

Plan to keep your survey open at least a month if possible in order to garner as many respondents as possible.

---

## Remember Your Sampling Lessons

How you distribute your survey can have far-reaching implications into the kind of data you collect and how much bias you are introducing to your data.  Although the most rigorous, and thus least biased, sampling strategy is randomly sampling, this is typically impractical when working with most companies, so you'll need to balance rigor with practicality.  Generally think through whether the channels for distribution you plan on using are more likely to pick up certain demographics or types of people.  For instance, if you are trying to entice senior citizens into taking your survey, requiring them to scan a QR code and participate on their phone may not be the best strategy. 

There is also trouble with a particular type of sampling strategy called *snowball sampling*.  Snowball sampling is when you ask respondents to forward your survey to other people they know.  While it can be helpful when you are trying to survey a small and hard-to-reach population, it should mostly be avoided, because you will bias your data a lot. If those who take the survey pass it on to friends, then your survey is inevitably getting distributed to people that have things in common with the original participant.  Maybe age, gender, race, same health conditions, physical proximity, etc. This means that you are artificially reducing the variation you find within your population and may end up with results wildly different than the actual population.

---

## Tips to Get Respondents

There are numerous research teams devoted to the best way to get survey respondents.  Here are some of the tips and tricks to increase that response rate just a little bit: 

* **Choose distribution times carefully:** Monday mornings are the best time to distribute; afternoons are the worst. 

* **Have a trusted authority figure distribute:** Ask a leader, manager, or other trustworthy authority figure to distribute the survey for you.  This must be someone known and respected by your survey respondents.  Further, the person distributing must not only be a reputable figure, but also *look* like a reputable sender.  This means using a company email address, sending things securely, etc. Without this, a lot of people will assume your survey is a scam, which is obviously detrimental. Avoid the spam filter at all costs!

* **Have a woman send it:** Especially in male-dominated fields, having a woman's name in the "from:" line may increase your response rate. 

* **Send reminders:** Send reminders at least every two weeks (you may opt for weekly reminders) at peak times (Monday mornings!) from a trusted leader.

* **Make survey invitations text based:** Often companies will block images, so ensure that key content is not contained within images and that things will still look good and make sense without rendered images.

* **Include a picture of the survey creator:** Even though your images may not be seen, when they are, having an image and some information about the survey creator can boost your response rates a little.  This is especially true (though unfair) if that person is attractive.

* **Explain why the survey is important:** Get respondents bought in, and share the lessons you've learned back with them. The more they feel important and needed, the higher your participation will be.

* **Include a progress bar:** Including a progress bar will allow participants to know that they are close to being done and may encourage them to finish the survey when they may not have otherwise.

---

### Offering Rewards

To reward or not reward? That 'tis the question! And a much debated question at that! While including a small incentive can increase your response rate, it may also change the motivation of your survey respondents.  They may be more focused on earning money, or on getting whatever the prize is, than answering the questions earnestly and honesty, which at best leads to mistakes on their part and at worse severely biases your survey results. If you do decide to risk using an incentive, it will typically be better to provide a very small reward to everyone who completes the survey, rather than to have people entered into a drawing for a larger prize.  This is particularly true when folks need to provide identifying information at the end in order to be eligible for the raffle; many people will not finish the survey with this point in mind. 

Whether you choose a reward for everyone or a drawing system, the incentive should not be so large as to be coercive in any way.  People should not feel like they HAVE to take the survey, just because of the reward. This is especially true for monetarily disadvantaged populations.  So, $1 to everyone would be appropriate; $50 to everyone would not be (even if you had the funds). 

Even if you want to incentivize survey participation, there may be some cases where it is not feasible.  Your company may not have or want to devote the funds that are required, and if you are doing something for the government, it is illegal to spend government money on incentives or other freebies and giveaways.  Knowing your left and right boundaries ahead of time is important! You never want to be in a situation where you have promised an incentive and are not able to provide one.

---

The option you WILL NOT want to choose is to send by email, as it will send a email to folks that doesn't come from a personalized sender and will often get trapped in spam mail, which is a great way to decrease your response rate, as no one will even see your survey link!

Once you click on the option to receive a web link, you'll get a page looking something like this one: 

![A survey monkey screen showing the web link for a survey. The survey creator can customize the link or copy it. The creator can also download a Q R code.](Media/survey23.png)

The link is now provided for you, and you can even download a QR code that can be put on print adds and then scanned into a mobile device, which is handy. Your survey is now OPEN, which means that anyone can take it provided they have the link.  It also means that you can't make changes to your survey easily anymore without changing data collection in the middle (never ideal), so for goodness sake, please proof far before you get to this step!

There are some options for survey distribution that you'll want to focus on.  Of course, SurveyMonkey puts the paid options on top, which you can ignore, but it's important to look at a few on the bottom:

![Survey monkey survey distribution options. Multiple responses, off. Response editing, on, until survey complete. Anonymous responses, off. Instant results, off.](Media/survey24.png)

```MULTIPLE OPTIONS``` decides whether participants can take your survey more than once. Typically, you don't want that to happen, so make sure that is ```off```. ```ANONYMOUS RESPONSES``` decides whether your participants will be anonymous.  By default, this is set to off, and so you get the IP address of the respondent.  However, you will typically want this to be ```On```! If you decide to have a non-anonymous survey, you need to disclose it immediately in your consent page.  Those are the two important ones, though you can also edit whether respondents get results and whether or not they can change their answers.  Use your best judgement as to whether you choose those options or not; it will depend on your survey goals.

## Examine Results

Lastly, click on the ```ANALYZE RESULTS``` tab to get a copy of your data and be able to look at the results from your survey.  Survey Monkey will automatically display for you some basic information like frequencies, but if you need to do in-depth analysis, it may not suit your needs.  Unfortunately, exporting your data in a useable format like a CSV file is a paid feature in Survey Monkey; you will need to discuss options with your supervisor should you be doing this with a company.

    </tr>

        <td>A way to gather information from others through written questions.</td>
