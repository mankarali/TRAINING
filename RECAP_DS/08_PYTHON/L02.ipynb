# Strings

You should recall from the last lesson that a _string_ is text &mdash; words, greetings, titles, names, etc. You saw how you can set a string, including using single and double quotes and special characters like tabs and newlines. Now, you will learn about a few operations that you can perform with strings, like combining and manipulating them.

----
## String Concatenation

The word *concatenation* means to combine. When concatenating strings, you use the `+` operator, similar to how you would add two numbers together. The difference, however, is that no arithmetic is performed. Instead, it simply glues the pieces together. Take a look at the example below:

```python
my_message = "Hello"
my_other_message = "how are you?"

# concatenate the above variables with a comma in between
my_final_message = my_message + ", " + my_other_message

print(my_final_message)
```

The output of the above print() statement would be: `"Hello, how are you?"` The value of the `my_message` variable is concatenated with `", "` and the value of the `my_other_message` variable using the `+` operator. The result of that concatenation is set as the value of the `my_final_message` variable, before it is printed.

Here's another example:

```python
day = "1"
year = "2017"
month = "August"

today_is = month + " " + day + ", " + year

print(today_is)
```

The `today_is` variable is assigned to the value of the `month`, `day`, and `year` variables concatenated together with separators. The output would be `August 1, 2017`.

# Converting Data to Strings

As a data scientist, there will be many times when you need to modify the type of variable you have! Below is a great example - what if some of your variables are set to integers, not strings, but you want to concatenate them? Well guess what? You can't! If you run the code below, you will see that you get an error:

```python
day = 1
year = 2017
month = "August"

today_is = month + " " + day + ", " + year

print(today_is)
```

The reason is because you cannot concatenate an integer (int) with a string. This is because the program will read the `+` sign as a `plus` sign rather than concatenation (you will be learning more about arithmetic operations with integers later in this lesson). 

---

## Using the str() Function

To handle this, you need to turn the integers into strings with the `str()` function, as shown below:

```python
day = 1
year = 2017
month = "August"

today_is = month + " " + str(day) + ", " + str(year)

print(today_is)
```
Now, when you run this, it will work. Pretty cool, right?

---
## Formatting Strings

Alternatively, instead of using `str()` on every piece of your concatenation, you could instead use the function `str.format()`. If you use it at the beginning, and in place of concatenation, then you can tell Python what to expect regarding the variables and tell it to format each as a string. For instance:

```python
day = 1
year = 2017
month = "August"

today_is = str.format("The date is {} {}, {} ", month, day, year)

print(today_is)
```

The output of the above will be:

```text
The date is August 1, 2017
```

---

### Variable Order Matters

The `str.format` takes in the string and within the string are empty curly braces `{}`. These curly braces are defined _after_ the string. The definitions of the curly braces are in the order you want the variables to be shown. So, in the above example, you want to see the month, day then year. If you change around those variables, the date will change:

```python
day = 1
year = 2017
month = "August"

today_is = str.format("The date is {}, {} {} ", year, month, day)

print(today_is)
```

Now the output will be:

```
The date is 2017, August 1
```

When you want to print a string using predefined variables, string concatenation and string formatting are both acceptable ways to accomplish this.

# String Operations

In Python, you can do so much more than just add two strings together. You can add or remove spacing, convert letters to lowercase or uppercase, and you can even replace one word with another.

---

## Making Text Strings Lowercase

The function `lower()` returns a string where all letters have been converted to lowercase.

The `lower()` function can be performed on a string to convert _all_ letters to lowercase. As you can see in the example below, a dot `.` and the `lower()` operation are appended to the string variable `my_message`.

```python
my_message = "THESE ARE ALL CAPS"

print(my_message.lower())
```

**OUTPUT**:

```
these are all caps
```

---
## Making Text Strings Uppercase

Great! You can also use `upper()` to do the opposite - it will return a string where all letters have been converted to uppercase. 

Similar to the `lower()` operation, the `upper()` operation can be called on strings. The result is that _all_ letters will be converted to uppercase.

```python
my_message = "these are all lower"

print(my_message.upper())
```

OUTPUT:

```
THESE ARE ALL LOWER
```

These operations can be handy in data wrangling, when you need to ensure that all your response options for a categorical variable are the same.  You don't want to count the frequencies of `cat` different from `CAT` or `Cat` or `cAt`, do you?

# Replacing Text in a String

You can use the function `.replace()` to return a string where you replace all instances of something with something else.  This function takes on the format `.replace('old', 'new')`, where all occurrences of `old` have been replaced by `new`.

If you need to replace some text in a string, you use the `replace` string operation. Unlike the previous two string operations, this one requires two extra inputs, called *;parameters*: one to specify the string occurrences that are to be changed, and one to specify what they should be changed to. These inputs are called _parameters_ and are related to _functions_ and _methods_, all of which you'll learn about soon!

This is easier to understand with examples.

```python
experience = "this game is fun"
better_experience = experience.replace("fun", "awesome")

print(experience)
print(better_experience)
```

**OUTPUT**:
```
this game is fun
this game is awesome
```

In the above example, the variable `better_experience` is set to the result of calling `.replace()` on the `experience` string variable to change _all_ occurrences of the word `fun` to the word `awesome`.

In the next example, more than one word is changed from a single call of the function `.replace()`. You can also see that single quotes are used instead of double quotes &mdash; both are acceptable.

```python
message = "I like cats, because cats are friendly"
new_message = message.replace('cats', 'dogs')

print(new_message)
```

**OUTPUT**:

```
I like dogs, because dogs are friendly
```

As with the `.upper()` and `.lower()` functions, `.replace()` will come in very handy when you start doing data wrangling! It can make it much easier to change data values if necessary.

# Splitting Text Strings Into Substrings

Using the `.split()` function, you can split a text string into substrings. A _substring_ is a portion of another string. The `.split()` operation has two optional parameters: one to specify how the string should be split (the _separator_), and another to specify the maximum number of splits that will be performed.

<div class="panel panel-success">
    <div class="panel-heading">
        <h3 class="panel-title">Fun Fact!</h3>
    </div>
    <div class="panel-body">
        <p>Splitting is often called "delimiting" as well, and when data is parsed into formats such as CSV, you are essentially using the split function to define each column by a separator!</p>
    </div>
</div>

---

## Splitting with Defaults

If no value is specified for the separator, it will default to splitting a string where it finds a _space_. If you do not specify a maximum number of splits, the default is -1, which means do all of them.

As always, this is best shown using examples:

```python
my_message = "Split these words for me"
split_at_spaces = my_message.split()

print(split_at_spaces)
```

**OUTPUT**:

The output is a list that has each word in its own set of quotes and separated by a comma. 

```python
['Split', 'these', 'words', 'for', 'me']
```

---

## Splitting with a Specified Separator

Now, an example where a separator is specified:

```python
my_message = "apples,oranges,bananas"
split_at_commas = my_message.split(",")

print(split_at_commas)
```

**OUTPUT**:
```
['apples', 'oranges', 'bananas']
```

Again, the output is a list.  But this time, instead of being split by spaces, which is the Python default, you have asked Python to split by commas, so each word before the comma becomes its own list object.

### Order of Mathematical Operations

In Python, as in traditional mathematics, an order of operations does exist. The expressions are not, truly, read from left to right. The operators will be evaluated in order of precedence. You can use the common technique for remembering the order of arithmetic operations with the abbreviation `PEMDAS`.

`PEMDAS` stands for:

1. <b>P</b>arentheses (simplify inside of them)
2. <b>E</b>xponents
3. <b>M</b>ultiplication and <b>D</b>ivision (from left to right)
4. <b>A</b>ddition and <b>S</b>ubtraction (from left to right)

If two operators share an operand the operator with the higher precedence goes first:

```python
a + b * c
```

The above expression would be treated like `a + (b * c)`, because the multiplication operator has the higher precedence.

Now, you can try the following operations by adding them to your `lesson_2.py` file and running them (remember, save the file, then right-click and select `Run Python File in Terminal` from the popup menu).

# Logical Operators

Earlier, you learned about comparison operators, such as `==,` `!=,` and <code>&lt;=</code>. The result of an operation using a comparison operator is a boolean; it either results in `True` or `False`. So, for instance, `person_age < 10` is an operation using the less than operator, and the result will be true if `person_age` is less than 10, otherwise false. But what happens if you need to test for multiple conditions? Just like multiple event probability, where you can make use of your boolean operators of "and", "or", and "not", you can use them in Python as well.  You'll start by working with `and`. 

---
## and

In Python, you can test for multiple conditions using a *logical operator*. For example, say you need to know whether someone can watch a movie. The movie is rated R, so they must be at least 18 years of age. They must also have enough money to pay for the ticket. Take a look at the code below, which combines these conditions:

```python
# details of the person who wants to watch the movie
person_age = 17
person_money = 25

# the requirements to watch the movie
age_restriction = 18
movie_price = 10

# conditions
is_old_enough = person_age >= age_restriction
has_enough_money = person_money >= movie_price

# two conditions combined using 'and'
can_watch_movie = is_old_enough and has_enough_money
print(can_watch_movie)

# the output is False, because the person's age is less than the required age
```

In the code above, `and` is used to combine the two conditions. The final check (`can_watch_movie`) basically says:

> The person can watch the movie if they're old enough AND they have enough money.

When using the `and` logical operator, you are not limited to only two conditions. For instance, this code adds a third condition: 

```python
# multiple conditions
result = (10 > 3) and (3 < 9) and ("this" != "that")
```

# Combining Logical Operators

When combining `and` and `or`, you may need to use parentheses in order to ensure that the operations are performed in the appropriate order. This is the order that Python will automatically perform your logical operators, unless you distinctly specify: 

* not
* and
* or

The `and` logical operator takes precedence over the `or` operator. The `not` operator takes precedence over the other two.

Consider an example where someone can watch a movie if:
- they are at least 18 years of age, and they have at least $15
- OR, they are a friend and have at least $5
- they must NOT have any food, regardless of the above

You can code this into Python in the following way:

```python
my_age = 17
my_money = 7
is_friend = True
has_food = False

can_watch_movie = not has_food and ((my_age >= 18 and my_money >= 15) or (is_friend and my_money >= 5))

print(can_watch_movie)
```

The first operation performed, above, is the `not has_food`. This condition must always hold true. If it is, then it confirms that either the person is old enough with enough money __OR__ they are a friend with a lower amount of money, but still enough.

If the first check (has_food) were `False`, the remainder of the expressions are not computed, because it would be a waste of time; they are skipped. It knows it will fail immediately.

# Decision-Making

As you write programs in Python, you will undoubtedly encounter situations that will require making decisions. Programming often involves checking a certain set of conditions and choosing an action based upon those conditions. This is done in code much like in real life. Consider the following example:

> A person can watch a movie __if__ they have $10

This can be translated to Python pretty easily:

```python
person_money = 12

if person_money >= 10:
  print("This person can watch the movie")
```

The output of the above code will be "This person can watch the movie." Why? Because they have $12, which is more than $10. The text is printed to the console only if the variable `person_money` has a value that is greater than or equal to 10, because you are only going to execute that command **if** the first statement is true.

What if you changed the value of `person_money` to 8, as shown below?

```python
person_money = 8

if person_money >= 10:
  print("This person can watch the movie")
```

Well, nothing gets printed to the console. But why?

The condition test of whether `person_money >= 10` fails, so the line to print is skipped. This is how `if` statements work. The thing only happens when the condition is true.

---

## General `if` syntax:

When using `if`, the syntax is the following:

```
if <condition>:
  <operations>
```

The above is generic code, so it won't run.  But it gives you the general format of how things should be written, and you will fill in the `<>` with your own conditions and operations. You will always start with the keyword `if`, and then follow that with the condition to test.  In the earlier example, the condition that was being tested was whether `person_money` was at least $10 or more.

Following the condition, you'll find a colon (`:`). A colon is required after the `if` and condition tests, or it won't run.  Python will be thinking that everything following is conditional, and will never recognize that it needs to stop and do the operation without that colon! You can think of the colon like a stopsign.  Without stop signs on the road, everyone would just continue driving forward forever without recognizing that there is a need to stop driving, look both ways, and then proceed in another direction. 

Remember that the block of operations to be performed __IF__ the condition is met (true) is indented to the right. Python will also be angry if you forget that indent; luckily VSC will indent for you automatically when you include the colon.  So get the first part right, and you should automatically get the second part right as well!

---

## else

In the example above, the print operation is only performed if the condition is true. So, what if you wanted to print something if the condition fails? Like to tell them "Sorry, but you don't have enough money." This is where the `else` keyword comes into play.

```python
person_money = 8

if person_money >= 10:
  print("This person can watch the movie")
else:
  print("Sorry, but you don't have enough money")
```

The output of the above code will be `"Sorry, but you don't have enough money"`, because `person_money` is __not__ greater than or equal to 10. It is something else.

So, both possible outcomes are handled above: they either have enough money, or they don't. Either way, the appropriate message is printed to the console.

---

### Syntax for else

The `else` statement is aligned with the `if` statement (the indentation is the same). As well, the operation that is performed for the `else` block is indented like for the `if` statement.

The `else` statement also requires the use of the colon after the keyword. 

---

## elif

Finally, if there are more than two possible outcomes, you can use the `elif` statement. Consider the following scenario:

> If your age is 18 or higher, you can watch R-rated movies. If your age is 13 or higher, you can watch PG13-rated movies. If your age is less than 13, you can only watch G-rated movies.

Below, you'll find the Python code to represent the above:

```python
person_age = 18

if person_age >= 18:
  print("I can watch an R-rated movie")
elif person_age >= 13:
  print("I can watch a PG13-rated movie")
else:
  print("I can watch a G-rated movie")
```

The `elif` statement is essentially a combination of two words: _else_ and _if_. Like an `if` statement, a condition test must follow the `elif` statement. And, like both `if` and `else`, a colon must reside at the end of the line for `elif` statements. Similarly, the operations must be indented like the other blocks.

The order in which you test the conditions can matter. The output from the above example is "I can watch an R-rated movie", which is what you would expect given the value of `person_age` (18). However, what happens if you change the order of condition tests?

```python
person_age = 18

if person_age >= 13:
  print("I can watch a PG13-rated movie")
elif person_age >= 18:
  print("I can watch an R-rated movie")
else:
  print("I can watch a G-rated movie")
```

The output of the above would be "I can watch a PG13-rated movie", which is different than the previous example. This is due to the change in sequence of the checks. Since the first condition (`person_age >= 13`) is true, it succeeds, printing out the message about the PG-13 movie. The remainder of the checks are skipped, because the first one was satisfied.

Now, try the following:

```python
person_age = 5

if person_age >= 18:
  print("I can watch an R-rated movie")
elif person_age >= 13:
  print("I can watch a PG13-rated movie")
else:
  print("I can watch a G-rated movie")
```

In the above code, the first test is whether the age is >= 18, which is false (5 < 18). So, it moves to the next test. However, this test fails, as well, since 5 is less than 13. Thus, the final `else` block is hit and the print operation is performed (there is no test for the final else).
