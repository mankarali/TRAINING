```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
names = list(iris.target_names)
#print(names)
print(f'x contient {x.shape[0]}exmples et {x.shape[1]} variables')
print(f'il y a {np.unique(y).size} classes')
```

## Graphique de classification avec plt.scatter

```python
#on fait passer 2 des 4 variables dans le scatter plot et couleurs les targets, aplha gère la transparence et s la taille des points
plt.scatter(x[:,0], x[:,1], c=y, alpha = 0.5,s=100) 
plt.xlabel('longeur du sepal')
plt.ylabel('largeur du sepal')
```

## Graphique 3D

```python
#avec mpl toolkit permet de representer plus que 2 variables
#%matplotlib
from mpl_toolkits.mplot3d import Axes3D
```

```python
#importer l'axe
ax = plt.axes(projection='3d')
ax.scatter(x[:,0],x[:,1],x[:,2] ,c = y)
plt.show()
```

```python
#voir des fonctions mathématiques
f = lambda x, y:np.sin(x)+ np.cos(x+y)
```

```python
#créer d'abord une grille avec numpy avec np.meshgrid
X = np.linspace(0, 5, 100) #absisse 
Y = np.linspace(0, 5, 100) # ordonnées
X, Y = np.meshgrid(X,Y)

Z = f(X,Y)
```

```python
Z.shape
```

```python
#visualisation 3D
ax = plt.axes(projection = '3d')
ax.plot_surface(X, Y, Z, cmap ='plasma' )
```

### Histogramme

```python
#histogramme longueur d'un sepal
#bins: c'est le nombre de section dans notre histogramme 
plt.hist(x[:,0], bins=30)
```

```python
# 2 histogrammes dans un même graphique 
plt.hist(x[:,0], bins = 20)
plt.hist(x[:,1], bins = 20)
```

```python
#histogramme 2 dimensions
plt.hist2d(x[:,0], x[:,1], cmap = 'Reds')
plt.xlabel('longueur sépal')
plt.ylabel('largeur sépal')
plt.colorbar()
```

```python
#analyser une image 
from scipy import misc #import image
face = misc.face(gray=True)
plt.imshow(face, cmap= 'gray')
```

```python
#analyser l'image en analysant la repartission des pixels
plt.hist(face.ravel(), bins=255)#ravel permet d'écraser le tableau
plt.show()
```

## les contours plots

```python
#permet une vue du dessus de nos fonctions permet de faire des optimisations
plt.contour(X, Y, Z,20, colors= 'black')
```

```python
#color map 
plt.contourf(X,Y,Z, 20, cmap = 'RdGy')
plt.colorbar()
```
